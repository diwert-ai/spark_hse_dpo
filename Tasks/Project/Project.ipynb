### `Описание проекта`

В качества финального проекта предлагается взять набор данных и выполнить стандартную последовательность действий для обучения модели машинного обучения с использованием библиотеки **Spark**. 

Более конкретно, примерный план выполнения проекта включает в себя следующие действия:

1. Скачать данные соревнования [M5 Forecasting](https://www.kaggle.com/competitions/m5-forecasting-accuracy/data)
2. Прочитать и разобрать описание соревнования на Kaggle. Выяснить, какие данные доступны и какой отклик предлагается предсказывать
3. Предложить формализацию задачи. Выбрать метрику качества модели и **обосновать выбор**
4. Провести Разведочный Анализ Данных (**EDA**). Выявить инсайты в данных, которые в дальнейшем можно использовать для построения модели
5. Выбрать модели для обучения. Сформировать подходящие признаковые пространства для выбранных моделей
6. Обучить модели
7. Оценить качество моделей на обучающей и отложенной выборках
8. Привести примеры успешных и неуспешных предсказаний. Провести анализ ошибок
9. Сделать **выводы**
10. Оформить **презентацию** с описанием проведённой работы

*Замечание:* Все действия над данными **должны** быть выполнены с использованием Spark. "Собирать" все данные на драйвер без предварительной агрегации и преобразования в Spark не рекомендуется, в том числе и при построении графиков и проведении EDA. 

*Общий совет:* при написании кода старайтесь реализовать максимально масштабируемое решение, которое будет работать, даже если объём данных значительно вырастет.
